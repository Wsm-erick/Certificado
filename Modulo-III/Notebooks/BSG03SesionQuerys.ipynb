```python
import pandas as pd
import sqlite3
from pathlib import Path

# Lista de archivos CSV a cargar
csv_files = ['BD_Negocios_uno.csv', 'BD_Negocios_Dos.csv']
table_names = ['negocios_uno', 'negocios_dos']  # Nombres de las tablas en SQLite

# Crear conexión a la base de datos SQLite
try:
    conn = sqlite3.connect('bd_negocio2.db')
    print("Conexión a la base de datos SQLite 'bd_negocio2.db' establecida con éxito.")

    # Iterar sobre los archivos CSV y sus respectivos nombres de tabla
    for csv_file, table_name in zip(csv_files, table_names):
        # Verificar si el archivo existe
        if not Path(csv_file).is_file():
            print(f"Error: El archivo '{csv_file}' no se encuentra en el directorio.")
            continue
        
        try:
            # Leer el archivo CSV con delimitador punto y coma
            df = pd.read_csv(csv_file, sep=';', encoding='latin1', on_bad_lines='skip')
            print(f"Archivo '{csv_file}' cargado exitosamente con {len(df)} registros.")

            # Convertir el DataFrame a una tabla SQLite
            df.to_sql(table_name, conn, if_exists='replace', index=False)
            print(f"Tabla '{table_name}' creada en la base de datos con éxito.")

        except Exception as e:
            print(f"Error al procesar el archivo '{csv_file}': {str(e)}")

    # Confirmar los cambios y cerrar la conexión
    conn.commit()
    print("Todos los datos han sido procesados y guardados en 'bd_negocio2.db'.")

except Exception as e:
    print(f"Error al conectar con la base de datos: {str(e)}")

finally:
    conn.close()
    print("Conexión a la base de datos cerrada.")
```
